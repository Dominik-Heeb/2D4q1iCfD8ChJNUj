## Conclusions <a class="anchor" id="Conclusions"></a>
[TOC](#TOC)

* There is no clear optimum for `k` and `bonus`.
* Using k = 1 implies using the bonus only.
* A combination k=2, bonus=0.5 might fit most scenarios.
* This result __depends on how we handle identical records__:
    * Accepting as referring to different candidates, as done here: conservative approach.
    * Deleting duplicates right at the beginning.
    * Apply bonus to duplicates too.
    * Include duplicates when calculating the metric (e.g. delta).
